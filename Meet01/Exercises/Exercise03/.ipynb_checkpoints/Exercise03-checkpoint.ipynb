## Exercise03 - Good Sleep

Your client, SleepInc, has shared anonymized sleep data from their hot new sleep tracking app SleepScope. As their data science consultant, your mission is to analyze the lifestyle survey data with Python to discover relationships between exercise, gender, occupation, and sleep quality. See if you can identify patterns leading to insights on sleep quality.

### ðŸ’¾ The data: sleep_health_data.csv

SleepInc has provided you with an anonymized dataset of sleep and lifestyle metrics for 374 individuals. This dataset contains average values for each person calculated over the past six months. The data is saved as `sleep_health_data.csv`.

The dataset includes 13 columns covering sleep duration, quality, disorders, exercise, stress, diet, demographics, and other factors related to sleep health. 

| Column | Description |
|---------|----------------------------------------|  
| `Person ID` | An identifier for each individual. |
| `Gender` | The gender of the person (Male/Female). |  
| `Age` | The age of the person in years. |
| `Occupation` | The occupation or profession of the person. |
| `Sleep Duration (hours)` | The average number of hours the person sleeps per day. |
| `Quality of Sleep (scale: 1-10)` | A subjective rating of the quality of sleep, ranging from 1 to 10. |
| `Physical Activity Level (minutes/day)` | The average number of minutes the person engages in physical activity daily. |  
| `Stress Level (scale: 1-10)` | A subjective rating of the stress level experienced by the person, ranging from 1 to 10. |
| `BMI Category` | The BMI category of the person (e.g., Underweight, Normal, Overweight). |
| `Blood Pressure (systolic/diastolic)` | The average blood pressure measurement of the person, indicated as systolic pressure over diastolic pressure. |
| `Heart Rate (bpm)` | The average resting heart rate of the person in beats per minute. |
| `Daily Steps` | The average number of steps the person takes per day. |
| `Sleep Disorder` | The presence or absence of a sleep disorder in the person (None, Insomnia, Sleep Apnea). |

### Tasks :
1. Which occupation has the lowest average sleep duration? 
2. Which occupation has the lowest average sleep quality?
3. Did the occupation with the lowest sleep duration also have the lowest sleep quality?
4. Calculate the ratio of each BMI Category that have been diagnosed with Insomnia?

```python
# Start coding here
# Use as many cells as you need
import pandas as pd
```

```python
sleep_health = pd.read_csv('sleep_health_data.csv')
```

```python
sleep_health.columns
```

```python
sleep_health.head()
```

```python
sleep_health.shape
```

### Answer 1 

```python
## this is series
## calculate the average sleep duration groupby Occupation
avg_sleep_per_occupations = sleep_health.groupby('Occupation')['Sleep Duration'].mean()

## find the minimum value in the series
shortest_avg_sleep = avg_sleep_per_occupations.min()
## find the occupation that hold the minimum value
occ_shortest_avg_sleep = avg_sleep_per_occupations.idxmin()
## print then
print(f'Shortest sleep is hold by {occ_shortest_avg_sleep} with average duration is {shortest_avg_sleep}')

# this is dataframe
avg_sleep_per_occupations = sleep_health.groupby('Occupation')['Sleep Duration'].mean().reset_index()
avg_sleep_per_occupations.columns = ['Occupation', 'Avg sleep']

avg_sleep_per_occupations

```

### Answer 2 

```python
## this is series
## calculate the average Quality of sleep groupby Occupation
avg_qosleep_per_occupations = sleep_health.groupby('Occupation')['Quality of Sleep'].mean()

## find the minimum value in the series
lowest_avg_qosleep = avg_qosleep_per_occupations.min()
## find the occupation that hold the minimum value
occ_lowest_avg_qosleep = avg_qosleep_per_occupations.idxmin()
## print then
print(f'lowest quality of sleep is hold by {occ_lowest_avg_qosleep} with average quality is {lowest_avg_qosleep}')

# this is dataframe
avg_qosleep_per_occupations = sleep_health.groupby('Occupation')['Quality of Sleep'].mean().reset_index()
avg_qosleep_per_occupations.columns = ['Occupation', 'Avg Quality of Sleep']
avg_qosleep_per_occupations['Avg Quality of Sleep'] = avg_qosleep_per_occupations['Avg Quality of Sleep'].round(2) 

avg_qosleep_per_occupations
```

### Answer 3

```python
import matplotlib.pyplot as plt

if occ_lowest_avg_qosleep == occ_shortest_avg_sleep:
    print(f'YES, {occ_shortest_avg_sleep} is lowest and shortest on Quality and Duration of Sleep')
else:
    print('NO')

result = avg_sleep_per_occupations.merge(avg_qosleep_per_occupations, on='Occupation')

display(result)

# Set up the figure and axes
fig, ax1 = plt.subplots(figsize=(20, 6))

# Plot avg_sleep
ax1.bar(result['Occupation'], result['Avg sleep'], color='blue', alpha=0.6, label='Avg Sleep')

# Plot avg_quality_of_sleep
ax2 = ax1.twinx()  # Create a second y-axis
ax2.plot(result['Occupation'], result['Avg Quality of Sleep'], color='red', marker='o', label='Avg Quality of Sleep')

# Title and labels
ax1.set_title('Average Sleep and Quality of Sleep by Occupation')
ax1.set_xlabel('Occupation')
ax1.set_ylabel('Avg Sleep', color='blue')
ax2.set_ylabel('Avg Quality of Sleep', color='red')

# Legends
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Show the plot
plt.show()

```

### Answer 4 

```python
# Filter rows where Sleep Disorder is Insomnia
insomnia = sleep_health[sleep_health['Sleep Disorder'] == 'Insomnia']

# count BMI category on insomnia dataframe
insomnia_counts = insomnia['BMI Category'].value_counts()

# count BMI Category one the entire dataset
total_counts = sleep_health['BMI Category'].value_counts()

# Calculate the ratio
ratios = insomnia_counts / total_counts

# create new dataframe of ratio
ratios_df = ratios.reset_index()
ratios_df.columns = ['BMI Category', 'Ratio']
ratios_df
```
