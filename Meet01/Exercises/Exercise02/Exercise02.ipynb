```python
## prompt I wrote on chat gpt
# assume I have pandas dataframe, 
# one of its header is age and 
# I want create new column age 
# that consists freqeuncy of range of age 
# '0-17', '18-25', '26-34', '35-44', '45-54', '55-64', '65+'

bins = [0, 17, 25, 34, 44, 54, 64, float('inf')]
labels = ['0-17', '18-25', '26-34', '35-44', '45-54', '55-64', '64+']

crimes['Vict Age Range'] = pd.cut(crimes['Vict Age'], bins=bins, labels=labels, right=True)

age_range_counts = crimes['Vict Age Range'].value_counts().sort_index()

crimes_age_range = pd.DataFrame(age_range_counts).reset_index()

crimes_age_range.columns = ['Vict Age Range', 'frequency']

crimes_age_range
```
