## Exercise01 - Public School Test Result

Every year, American high school students take SATs, which are standardized tests intended to measure literacy, numeracy, and writing skills. There are three sections - reading, math, and writing, each with a **maximum score of 800 points**. These tests are extremely important for students and colleges, as they play a pivotal role in the admissions process.

Analyzing the performance of schools is important for a variety of stakeholders, including policy and education professionals, researchers, government, and even parents considering which school their children should attend. 

You have been provided with a dataset called `schools.csv`, which is previewed below.

You have been tasked with answering three key questions about New York City (NYC) public school SAT performance.

### Tasks :
1. Which NYC schools have the best math results? 
   - The best math results are at least 80% of the *maximum possible score of 800* for math.
   - Save your results in a pandas DataFrame, including "school_name" and "average_math" columns, sorted by "average_math" in descending order.
2. What are the top 10 performing schools based on the combined SAT scores?
   - Save your results as a pandas DataFrame containing the "school_name" and a new column named "total_SAT", with results ordered by "total_SAT" in descending order.
3. Which single borough has the largest standard deviation in the combined SAT score?
   - Save your results as a pandas DataFrame.
   - The DataFrame should contain one row, with:
      - "borough" - the name of the NYC borough with the largest standard deviation of "total_SAT".
      - "num_schools" - the number of schools in the borough.
      - "average_SAT" - the mean of "total_SAT".
      - "std_SAT" - the standard deviation of "total_SAT".
   - Round all numeric values to two decimal places.

```python
# Re-run this cell 
import pandas as pd

# Read in the data
schools = pd.read_csv("schools.csv")

# Preview the data
schools.head()
```

## Answer 1 

```python
## took best schools with scores more than 80% of 800 or 80x8
best_math_schools = schools[schools.average_math >= (80*8)]
## too only 2 columns and arrange the order descendingly
bms = best_math_schools[['school_name', 'average_math']].sort_values(by='average_math', ascending=False)
## reset the index
bms.reset_index(drop=True, inplace=True)
## start the index from 1
bms.index = bms.index+1
## display
bms
```

## Answer 2

```python
## add total_SAT
# schools = schools.assign(total_SAT= schools['average_math'] + schools['average_reading'] + schools['average_writing'])
schools = schools.assign(total_SAT= schools.average_math + schools.average_reading + schools.average_writing)
## take 10 highest value of total_SAT descendingly
top_10_schools = schools.nlargest(10, 'total_SAT')[['school_name', 'total_SAT']].sort_values(by='total_SAT', ascending=False)
## reset the index
top_10_schools.reset_index(drop=True, inplace=True)
## start index with 1
top_10_schools.index = top_10_schools.index + 1
## display
top_10_schools
```

## Answer 3

```python
schools.head()
```

```python
## calculate total school for each city
num_schools = schools.borough.value_counts().reset_index()
num_schools.columns = ['borough', 'num_schools']

## calculate average of total_SAT
average_SAT = schools.groupby('borough')['total_SAT'].mean().reset_index()
average_SAT.columns = ['borough', 'average_SAT']

## calculate std
std_SAT = schools.groupby('borough')['total_SAT'].std().reset_index()
std_SAT.columns = ['borough', 'std_SAT']

## merge
result = num_schools.merge(average_SAT, on='borough')
result = result.merge(std_SAT, on='borough')

## alternative
#result_a = schools.groupby('borough').agg(num_schools=('borough', 'size'), average_SAT=('borough', 'mean'), std_SAT=('borough', 'std')).reset_index()

## rounded to two decimal places
result.average_SAT = result.average_SAT.round(2)
result.std_SAT = result.std_SAT.round(2)

result
```
